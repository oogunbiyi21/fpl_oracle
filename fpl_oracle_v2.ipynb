```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Fetch Data from the FPL API

```python
def get_fpl_data(endpoint):
    url = f'https://fantasy.premierleague.com/api/{endpoint}/'
    response = requests.get(url)
    response.raise_for_status()
    return response.json()
```

## Data prep

```python
 # Fetch data
bootstrap_data = get_fpl_data('bootstrap-static')
elements_data = bootstrap_data['elements']
teams_data = bootstrap_data['teams']

positions_data = bootstrap_data['element_types']

# Convert to DataFrames
players_df = pd.DataFrame(elements_data)
teams_df = pd.DataFrame(teams_data)
positions_df = pd.DataFrame(positions_data)

# Data clean
players_df = players_df[['id', 'web_name', 'team', 'element_type', 'now_cost', 'total_points', 'minutes', 'points_per_game', 'value_season', 'status']]
players_df['now_cost'] = players_df['now_cost'] / 10  # Convert cost to millions
players_df['value_season'] = pd.to_numeric(players_df['value_season'], errors='coerce')

# Filter out injured or suspended players
available_players_df = players_df[players_df['status'] == 'a']

# Calculate ROI (Fantasy points per million)
available_players_df['roi'] = available_players_df['total_points'] / (available_players_df['now_cost'] / 10)

# Create position mapping
position_mapping = positions_df.set_index('id')['singular_name_short'].to_dict()
```

## Team selection algorithm

```python
def select_fantasy_team(available_players_df, teams_data, position_mapping):
    BUDGET = 100  # 100 million
    SQUAD_SIZE = 15
    POSITION_LIMITS = {'GKP': 2, 'DEF': 5, 'MID': 5, 'FWD': 3}
    TEAM_LIMIT = 3

    # Initialize the team
    team = []
    budget_left = BUDGET
    position_count = {pos: 0 for pos in POSITION_LIMITS.keys()}
    team_count = {team['id']: 0 for team in teams_data}

    # Add position to the players DataFrame
    available_players_df['position'] = available_players_df['element_type'].map(position_mapping)

    # Select top 3 star players by total points first
    star_players = available_players_df.nlargest(3, 'total_points')
    for _, player in star_players.iterrows():
        team.append(player)
        budget_left -= player['now_cost']
        position_count[player['position']] += 1
        team_count[player['team']] += 1

    # Remove selected star players from available players
    available_players_df = available_players_df.drop(star_players.index)

    # Fill the remaining spots with high ROI players
    for _, player in available_players_df.sort_values('roi', ascending=False).iterrows():
        if len(team) >= SQUAD_SIZE:
            break
        if budget_left < player['now_cost']:
            continue
        if position_count[player['position']] >= POSITION_LIMITS[player['position']]:
            continue
        if team_count[player['team']] >= TEAM_LIMIT:
            continue
        
        team.append(player)
        budget_left -= player['now_cost']
        position_count[player['position']] += 1
        team_count[player['team']] += 1

    # Convert the team to DataFrame for display
    team_df = pd.DataFrame(team)
    return team_df
```

```python
# Function to plot fixture difficulty for a chosen gameweek
def plot_fixture_difficulty(gameweek):
    # Filter for the chosen gameweek
    gw_fixtures_df = fixtures_df[fixtures_df['event'] == gameweek]
    
    if gw_fixtures_df.empty:
        print(f"No fixtures found for gameweek {gameweek}.")
        return
    
    # Plot the fixture difficulty for the chosen gameweek
    plt.figure(figsize=(14, 8))
    
    # Home teams
    sns.barplot(
        x='team_h',
        y='team_h_difficulty',
        data=gw_fixtures_df,
        color='blue',
        label='Home Team Difficulty'
    )
    
    # Away teams
    sns.barplot(
        x='team_a',
        y='team_a_difficulty',
        data=gw_fixtures_df,
        color='red',
        label='Away Team Difficulty'
    )
    
    plt.xlabel('Team')
    plt.ylabel('Difficulty')
    plt.title(f'Fixture Difficulty for Gameweek {gameweek}')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# Example usage: Plot fixture difficulty for gameweek 5
plot_fixture_difficulty(1)
```

## Accounting for fixture difficulty in team selection

```python
def calculate_player_fixture_difficulty(team_id, fixtures_df, gw_start, gw_end):
    # Find all fixtures for the player's team
    gw_list = [gw for gw in range(gw_start, gw_end+1)]
    limited_fixtures_df = fixtures_df[fixtures_df["event"].isin(gw_list)]
    team_fixtures = fixtures_df[(fixtures_df['team_h'] == team_id) | (fixtures_df['team_a'] == team_id)]
    
    # Calculate the average difficulty rating
    team_fixtures['difficulty'] = team_fixtures.apply(
        lambda x: x['team_h_difficulty'] if x['team_h'] == team_id else x['team_a_difficulty'],
        axis=1
    )
    average_difficulty = team_fixtures['difficulty'].mean()
    return average_difficulty

# available_players_df['fixture_difficulty'] = available_players_df['team'].apply(lambda x: calculate_player_fixture_difficulty(x, fixtures_df))
```

```python
# Adjust player selection based on points per game and fixture difficulty
def rank_players(available_players_df):
    # Higher points_per_game and lower fixture_difficulty are better
    available_players_df['rank_score'] = available_players_df['roi'] / available_players_df['fixture_difficulty']
    return available_players_df.sort_values(by='rank_score', ascending=False)

# ranked_players_df = rank_players(available_players_df)
```
